```python
"""
FER2013 Facial Emotion Recognition - CNN Classifier
Classes: Angry, Fear, Happy, Neutral, Sad, Surprise (Disgust excluded)
Optimized for Kaggle Notebooks
"""

import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.metrics import classification_report, confusion_matrix

# ─────────────────────────────
# KAGGLE PATHS
# In Kaggle, the dataset is auto-mounted at /kaggle/input/
# ─────────────────────────────
BASE_DIR  = "/kaggle/input/datasets/msambare/fer2013"
TRAIN_DIR = os.path.join(BASE_DIR, "train")
TEST_DIR  = os.path.join(BASE_DIR, "test")

# Verify paths
print("Train path exists:", os.path.exists(TRAIN_DIR))
print("Test path exists :", os.path.exists(TEST_DIR))
print("Classes found    :", os.listdir(TRAIN_DIR))

# ─────────────────────────────
# CONFIG
# ─────────────────────────────
IMG_SIZE   = (48, 48)
BATCH_SIZE = 64
EPOCHS     = 50
LR         = 1e-3
CLASSES    = ["angry", "fear", "happy", "neutral", "sad", "surprise"]  # disgust excluded

# ─────────────────────────────
# DATA LOADERS
# ─────────────────────────────
train_gen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True,
    zoom_range=0.1
)

test_gen = ImageDataGenerator(rescale=1./255)

train_data = train_gen.flow_from_directory(
    TRAIN_DIR,
    target_size=IMG_SIZE,
    color_mode="grayscale",
    batch_size=BATCH_SIZE,
    classes=CLASSES,           # disgust folder is automatically ignored
    class_mode="categorical",
    shuffle=True
)

test_data = test_gen.flow_from_directory(
    TEST_DIR,
    target_size=IMG_SIZE,
    color_mode="grayscale",
    batch_size=BATCH_SIZE,
    classes=CLASSES,
    class_mode="categorical",
    shuffle=False
)

print(f"\nClass indices : {train_data.class_indices}")
print(f"Train samples : {train_data.samples}")
print(f"Test  samples : {test_data.samples}")

# ─────────────────────────────
# MODEL
# ─────────────────────────────
def build_model(num_classes=6):
    model = models.Sequential([
        # Block 1
        layers.Conv2D(32, (3,3), padding="same", activation="relu", input_shape=(48, 48, 1)),
        layers.BatchNormalization(),
        layers.Conv2D(32, (3,3), padding="same", activation="relu"),
        layers.BatchNormalization(),
        layers.MaxPooling2D(2, 2),
        layers.Dropout(0.25),

        # Block 2
        layers.Conv2D(64, (3,3), padding="same", activation="relu"),
        layers.BatchNormalization(),
        layers.Conv2D(64, (3,3), padding="same", activation="relu"),
        layers.BatchNormalization(),
        layers.MaxPooling2D(2, 2),
        layers.Dropout(0.25),

        # Block 3
        layers.Conv2D(128, (3,3), padding="same", activation="relu"),
        layers.BatchNormalization(),
        layers.Conv2D(128, (3,3), padding="same", activation="relu"),
        layers.BatchNormalization(),
        layers.MaxPooling2D(2, 2),
        layers.Dropout(0.25),

        # Classifier head
        layers.Flatten(),
        layers.Dense(256, activation="relu"),
        layers.BatchNormalization(),
        layers.Dropout(0.5),
        layers.Dense(num_classes, activation="softmax")
    ])
    return model

model = build_model()
model.summary()

model.compile(
    optimizer=Adam(learning_rate=LR),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

callbacks_list = [
    EarlyStopping(monitor="val_loss", patience=8, restore_best_weights=True),
    ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=4, min_lr=1e-6, verbose=1)
]

# ─────────────────────────────
# TRAIN
# ─────────────────────────────
history = model.fit(
    train_data,
    validation_data=test_data,
    epochs=EPOCHS,
    callbacks=callbacks_list
)

# ─────────────────────────────
# EVALUATE
# ─────────────────────────────
loss, acc = model.evaluate(test_data)
print(f"\nTest Accuracy: {acc*100:.2f}%  |  Test Loss: {loss:.4f}")

y_true = test_data.classes
y_pred = np.argmax(model.predict(test_data), axis=1)

print("\nClassification Report:")
print(classification_report(y_true, y_pred, target_names=CLASSES))

# ─────────────────────────────
# PLOTS
# ─────────────────────────────
# Accuracy & Loss curves
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].plot(history.history["accuracy"], label="Train")
axes[0].plot(history.history["val_accuracy"], label="Val")
axes[0].set_title("Accuracy"); axes[0].legend()

axes[1].plot(history.history["loss"], label="Train")
axes[1].plot(history.history["val_loss"], label="Val")
axes[1].set_title("Loss"); axes[1].legend()
plt.tight_layout(); plt.show()

# Confusion Matrix
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=CLASSES, yticklabels=CLASSES)
plt.title("Confusion Matrix")
plt.ylabel("True Label"); plt.xlabel("Predicted Label")
plt.tight_layout(); plt.show()

# ─────────────────────────────
# SAVE MODEL (Kaggle output dir)
# ─────────────────────────────
model.save("/kaggle/working/fer2013_emotion_model.h5")
print("Model saved to /kaggle/working/fer2013_emotion_model.h5")
```

```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model
import matplotlib.pyplot as plt

# ---- Paths ----
model_path = "/kaggle/working/fer2013_emotion_model.h5"
image_path = "/kaggle/input/datasets/ambatiakkhashsaye/image2-data/sad image.jpg"

# ---- Emotion labels (without disgust) ----
emotions = ['angry','fear','happy','sad','surprise','neutral']

# ---- Load model ----
model = load_model(model_path)

# ---- Read and preprocess image ----
img = cv2.imread(image_path)
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_gray = cv2.resize(img_gray, (48,48))
img_gray = img_gray / 255.0
img_gray = np.reshape(img_gray,(1,48,48,1))

# ---- Predict ----
pred = model.predict(img_gray)
emotion = emotions[np.argmax(pred)-1]

# ---- Show result ----
print("Predicted Emotion:", emotion)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Prediction: " + emotion)
plt.axis("off")
plt.show()
```
